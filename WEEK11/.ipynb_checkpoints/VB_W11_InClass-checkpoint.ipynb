#### a) Import data and put into df: I choose bitcoin searches worldwide

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

df = pd.read_csv("gtrend_bitcoin.csv", parse_dates=True)
df.head()
```

```python
df.index.names = ['Month']
df = df.rename(columns={'Category: All categories': 'Bitcoin search numbers'})
df.replace(to_replace = '<1', value = '1', inplace = True)
df1 = df.iloc[70 : , :] # no interest in bitcoin at the beginning: all zero, these are relative frequencies
df2 = pd.to_numeric(df1['Bitcoin search numbers']).to_frame()
```

### 2) Fourier transform:

```python
x = np.arange(1,200,1)
y = np.cos(x/2)

_ = plt.figure( figsize = (15,4))
_ = plt.plot(x,y)
_ = plt.xlabel('Time')
_ = plt.ylabel('Intensity')
plt.show()
```

```python
#power spectral density
psd = np.abs(np.fft.rfft(y))
freqs = np.fft.rfftfreq(len(y))
plt.plot(freqs,psd)

#amplitude
plt.axvline(1/(4*np.pi),c='y')
plt.xlabel('Frequency')
plt.ylabel('Intensity')
plt.show()
```

#### Brief explanation of Fourier transform: in the above example the Fourier transform renders a representation of a function in the frequency domain (while the original function acted in the time domain). The concept of fourier transform is more general, though: all periodic functions inhabit a function space which has eigenfunctions of the form exp(j x frequency x time), where freqencies vary depending on the period. Thus all such functions can be expressed as a linear combination of the mentioned eigenfunctions. Just like eigenvectors of vectorspaces, eigenfunctions of function spaces are orthonormal and when applied the scalar product (which in this case is integration of the product of functions over the period), the coefficients can be extracted. In the case of non-periodic functions this does not work, but the process can be generalized and this is what we call a Fourier transformation which in this general case instead of coefficients results in a function.
