- Types of APIs:
- 4 types of APIs:

- 1.Public API: A Public API is open to everyone without restrictions

- 2.Partner APIs: A licence is necessary in order to access the API. Prevalent in business activities.

- 3.Private APIs: For internal use within a company.

- 4.Composite APIs: Two or more APIs are combined to form a sequence of operations. Useful in optimalization for speed and performance
- 
- Why are APIs important?
- APIs (Application Programming Interfaces)are important for at east two reasons: because they allow access to data from applicationsin an organized manner and they make it easy to add new functionalities to any application or program without the need to recreate the code every time.
